# Data Preprocessing

```python
X = df.drop('quality',axis=1)
```

```python
X
```

```python
Y = df['quality'].apply(lambda y_value : 1 if y_value>=6 else 0) #1--good quality wien , 0 -->bad quality wine
```

```python
Y
```

```python
Y.value_counts() #almost 50% of both good and bad quality wine is present 
```

# Training and Testing 


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 3)
```

```python
X.shape , X_test.shape, X_test.shape
```

# Model training

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
wine_quality_predictor = RandomForestClassifier() #random forest classifier algorithm is applied on our model
```

```python
#training the model 
wine_quality_predictor.fit(X_train,Y_train) 
```

# Model testing and Evaluation 

```python
wine_quality_prediction = wine_quality_predictor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
model_accuracy = accuracy_score(wine_quality_prediction,Y_test)
```

```python
model_accuracy
```
